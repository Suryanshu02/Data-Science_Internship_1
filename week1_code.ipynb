```python
# Basic Data Processing Script
import pandas as pd

# -------------------------------
# 1. Load Dataset
# -------------------------------
# Replace with your file name
df = pd.read_csv('StudentsPerformance.csv')

print("Original Data:")
print(df.head())

# -------------------------------
# 2. Basic Processing
# -------------------------------

# Example: If dataset has 'math score' column
if 'math score' in df.columns:

    # Calculate average
    avg_marks = df['math score'].mean()

    # Maximum and Minimum
    max_marks = df['math score'].max()
    min_marks = df['math score'].min()

    print("\nStatistics:")
    print("Average Marks:", avg_marks)
    print("Maximum Marks:", max_marks)
    print("Minimum Marks:", min_marks)

# -------------------------------
# 3. Add New Column
# -------------------------------
# Example: Grade system
def get_grade(math_score):
    if math_score >= 90:
        return 'A'
    elif math_score >= 75:
        return 'B'
    elif math_score >= 50:
        return 'C'
    else:
        return 'Fail'

if 'math score' in df.columns:
    df['grade'] = df['math score'].apply(get_grade)

# -------------------------------
# 4. Grouping Data
# -------------------------------
if 'grade' in df.columns:
    summary = df.groupby('grade')['math score'].mean()
    print("\nAverage Marks by Grade:")
    print(summary)

# -------------------------------
# 5. Save Processed Data
# -------------------------------
df.to_csv('processed_data.csv', index=False)

print("\nProcessed Data Saved Successfully!")
```

```python
# Rename the 'grade' column to 'math\'s grade'
if 'grade' in df.columns:
    df = df.rename(columns={'grade': 'math\'s grade'})
    print("Column 'grade' renamed to 'math\'s grade' successfully.")
    display(df.head())
else:
    print("The 'grade' column does not exist in the DataFrame.")
```

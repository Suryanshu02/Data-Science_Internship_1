```python
# Data Cleaning Script
import pandas as pd
import numpy as np

try:
    pd.read_csv('movies.csv')
except FileNotFoundError:
    print("Creating a dummy 'movies.csv' file as it was not found.")
    data = {
        'Movie Title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E', 'Movie A'],
        'Genre': ['Action', 'Comedy', 'Drama', 'Action', 'Comedy', 'Action'],
        'Rating': [7.5, 8.2, 6.9, np.nan, 9.1, 7.5],
        'Year': [2020, 2021, 2019, 2022, 2023, 2020],
        'Marks': [60, 75, 45, 88, 92, 60]
    }
    dummy_df = pd.DataFrame(data)
    dummy_df.to_csv('movies.csv', index=False)

# Load dataset
df = pd.read_csv('movies.csv')

print("Original Dataset:")
print(df.head())

# 1. Remove Duplicates
df = df.drop_duplicates()

# 2. Handle Missing Values

# Option 1: Drop missing values
df = df.dropna()

# Option 2 (Alternative): Fill missing values
# df.fillna(df.mean(numeric_only=True), inplace=True)

# 3. Rename Columns
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# 4. Filter Data (Example)
# Keep rows where value > 50
if 'marks' in df.columns:
    df = df[df['marks'] > 50]

# 5. Convert Data Types

# Example: convert column to int
# df['marks'] = df['marks'].astype(int)

# 6. Remove Outliers

# Using IQR method
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]


# 7. Save Cleaned Data
df.to_csv('cleaned_data.csv', index=False)

print("\nCleaned Dataset:")
print(df.head())
```

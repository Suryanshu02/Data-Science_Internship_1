```python
# Week 3 Project: Data Cleaning and Analysis
import pandas as pd
import numpy as np

# 1. Create Sample Dataset
data = {
    'Name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Marks': [80, 90, np.nan, 70, 60, np.nan],
    'Age': [20, 21, 19, np.nan, 22, 20]
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)

# 2. Handle Missing Values
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)

# 3. Basic Analysis
avg_marks = df['Marks'].mean()
max_marks = df['Marks'].max()
min_marks = df['Marks'].min()

print("\nStatistics:")
print("Average Marks:", avg_marks)
print("Max Marks:", max_marks)
print("Min Marks:", min_marks)

# 4. NumPy Operations
marks_array = np.array(df['Marks'])
print("\nNumPy Array:", marks_array)
print("Squared Marks:", marks_array ** 2)

# 5. Grouping
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')
grouped = df.groupby('Result')['Marks'].mean()

print("\nAverage Marks by Result:\n", grouped)

# 6. Save Clean Data
df.to_csv('week3_cleaned_data.csv', index=False)

print("\nCleaned Data Saved Successfully!")
```
